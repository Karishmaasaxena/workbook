````python
# i
int_list = list(range(90, 131))
print("list: ",int_list)

def mean(int_list):
    return sum(int_list) / len(int_list)

print("mean: ",mean(int_list))
def median(int_list):
    sorted_list = sorted(int_list)
    n = len(sorted_list)
    if n % 2 == 0:
        return (sorted_list[n//2 - 1] + sorted_list[n//2]) / 2
    else:
        return sorted_list[n//2]

print("median: ",median(int_list))
````

````python
# ii
def mode(int_list):
    freq = {}
    for num in int_list:
        if num in freq:
            freq[num] += 1
        else:
            freq[num] = 1
    max_freq = max(freq.values())
    modes = [num for num, f in freq.items() if f == max_freq]
    return modes

print("mode: ",mode(int_list))
````

````python
# iii
def weighted_mean(values, weights):
    weighted_sum = sum(value * weight for value, weight in zip(values, weights))
    total_weight = sum(weights)
    return weighted_sum / total_weight

values = [1, 2, 3, 4, 5]
weights = [2, 1, 2, 1, 2]
print("weighted mean: ",weighted_mean(values, weights))
````

````python
# iv
def geometric_mean(numbers):
    product = 1
    for num in numbers:
        product *= num
    return product ** (1/len(numbers))

print("geometric mean: ",geometric_mean(int_list))
````

````python
# v
def harmonic_mean(numbers):
    reciprocals = [1/num for num in numbers]
    return len(numbers) / sum(reciprocals)

print("harmonic mean: ",harmonic_mean(int_list))
````

````python
# vi
def midrange(numbers):
    return (min(numbers) + max(numbers)) / 2

print("midrange: ",midrange(int_list))
````

````python
# vii
def trimmed_mean(numbers, trim_percentage):
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)
    num_to_remove = int(n * trim_percentage / 100)
    trimmed_numbers = sorted_numbers[num_to_remove:-num_to_remove]

    return sum(trimmed_numbers) / len(trimmed_numbers)
print("trimmed mean: ",trimmed_mean(int_list, 10))
````

)  Write a Python class representing a discrete random variable with methods to calculate its expected
value and variance

````python
class DiscreteRandomVariable:
    def __init__(self, values, probabilities):
        if len(values) != len(probabilities):
            raise ValueError("Values and probabilities lists must have the same length.")

        self.values = values
        self.probabilities = probabilities

        if not all(0 <= p <= 1 for p in probabilities):
            raise ValueError("Probabilities must be between 0 and 1.")

        if abs(sum(probabilities) - 1) > 1e-6:  # Allow for small floating-point errors
            raise ValueError("Probabilities must sum to 1.")

    def expected_value(self):
        return sum(value * prob for value, prob in zip(self.values, self.probabilities))

    def variance(self):
        expected_value = self.expected_value()
        return sum((value - expected_value) ** 2 * prob for value, prob in zip(self.values, self.probabilities))
````

0) Implement a program to simulate the rolling of a fair six-sided die and calculate the expected value and
variance of the outcomes.

````python
import random

def simulate_die_roll(num_trials):
    outcomes = []
    for _ in range(num_trials):
        outcomes.append(random.randint(1, 6))
    return outcomes

def calculate_expected_value(outcomes):
    return sum(outcomes) / len(outcomes)

def calculate_variance(outcomes):
    mean = calculate_expected_value(outcomes)
    squared_differences = [(x - mean) ** 2 for x in outcomes]
    return sum(squared_differences) / len(outcomes)

# Simulate 10,000 die rolls
num_trials = 10000
outcomes = simulate_die_roll(num_trials)

# Calculate expected value and variance
expected_value = calculate_expected_value(outcomes)
variance = calculate_variance(outcomes)

print(f"Expected Value: {expected_value}")
print(f"Variance: {variance}")
````

Create a Python function to generate random samples from a given probability distribution (e.g.,
binomial, Poisson) and calculate their mean and variance.

````python
import random
from scipy.stats import binom, poisson

def generate_samples(distribution, params, size):
    """Generates random samples from a given probability distribution.

    Args:
      distribution: The name of the distribution (e.g., 'binomial', 'poisson').
      params: A dictionary of parameters for the distribution.
      size: The number of samples to generate.

    Returns:
      A NumPy array of random samples.
    """

    if distribution == 'binomial':
        n = params['n']
        p = params['p']
        return binom.rvs(n, p, size=size)
    elif distribution == 'poisson':
        lam = params['lam']
        return poisson.rvs(lam, size=size)
    else:
        raise ValueError("Unsupported distribution.")


def calculate_mean_variance(samples):
    """Calculates the mean and variance of a list of samples.

    Args:
      samples: A NumPy array of samples.

    Returns:
      A tuple of mean and variance.
    """
    mean = sum(samples) / len(samples)
    variance = sum([(x - mean) ** 2 for x in samples]) / len(samples)
    return mean, variance

# Example usage
distribution = 'binomial'
params = {'n': 10, 'p': 0.5}  # Example parameters for binomial distribution
size = 1000  # Number of samples to generate
samples = generate_samples(distribution, params, size)
mean, variance = calculate_mean_variance(samples)

print(f"Mean: {mean}")
print(f"Variance: {variance}")


distribution = 'poisson'
params = {'lam': 5}  # Example parameters for poisson distribution
size = 1000  # Number of samples to generate
samples = generate_samples(distribution, params, size)
mean, variance = calculate_mean_variance(samples)

print(f"Mean: {mean}")
print(f"Variance: {variance}")
````

 Write a Python script to generate random numbers from a Gaussian (normal) distribution and compute
the mean, variance, and standard deviation of the samples.

````python
import numpy as np

def gaussian_stats(mu, sigma, num_samples):
    """Generates random numbers from a Gaussian distribution and computes statistics.

    Args:
        mu: Mean of the Gaussian distribution.
        sigma: Standard deviation of the Gaussian distribution.
        num_samples: Number of random samples to generate.

    Returns:
        A tuple containing the mean, variance, and standard deviation of the samples.
    """
    samples = np.random.normal(loc=mu, scale=sigma, size=num_samples)
    mean = np.mean(samples)
    variance = np.var(samples)
    std_dev = np.std(samples)
    return mean, variance, std_dev

# Example usage:
mu = 0  # Mean
sigma = 1  # Standard deviation
num_samples = 1000  # Number of samples

mean, variance, std_dev = gaussian_stats(mu, sigma, num_samples)

print(f"Mean: {mean}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_dev}")
````

  (iii) Write a function that calculates the covariance between two columns.


  (iv) Implement a Python program that calculates the Pearson correlation coefficient between two columns.


  (v) Write a script to visualize the correlation between two specific columns in a Pandas DataFrame using
scatter plots.

````python
import seaborn as sns
import pandas as pd
import numpy as np

# Load the tips dataset
tips = sns.load_dataset('tips')

# (i) Function to calculate skewness
def calculate_skewness(data):
    return data.skew()

# (ii) Program to determine skewness type
def analyze_skewness(data):
    skewness = calculate_skewness(data)
    if skewness > 0.5:
        print("Positive skewness")
    elif skewness < -0.5:
        print("Negative skewness")
    else:
        print("Approximately symmetric")

# (iii) Function to calculate covariance
def calculate_covariance(data1, data2):
    return np.cov(data1, data2)[0, 1]

# (iv) Program to calculate Pearson correlation coefficient
def calculate_pearson_correlation(data1, data2):
    return np.corrcoef(data1, data2)[0, 1]

# (v) Script to visualize correlation using scatter plots
def visualize_correlation(data1, data2):
    import matplotlib.pyplot as plt
    plt.scatter(data1, data2)
    plt.xlabel("total_bill")
    plt.ylabel("tip")
    plt.title('Correlation between total_bill and tip')
    plt.show()

# Calculate and print skewness
print("Skewness of total_bill:", calculate_skewness(tips['total_bill']))
print("Skewness of tip:", calculate_skewness(tips['tip']))

# Analyze skewness
print("Analysis for total_bill:")
analyze_skewness(tips['total_bill'])
print("Analysis for tip:")
analyze_skewness(tips['tip'])

# Calculate and print covariance
covariance = calculate_covariance(tips['total_bill'], tips['tip'])
print("Covariance between total_bill and tip:", covariance)

# Calculate and print Pearson correlation coefficient
correlation = calculate_pearson_correlation(tips['total_bill'], tips['tip'])
print("Pearson correlation coefficient between total_bill and tip:", correlation)

# Visualize correlation with scatter plot
visualize_correlation(tips['total_bill'], tips['tip'])
````

 Write a Python function to calculate the probability mass function (PMF) of Poisson distribution.

````python
from scipy.stats import poisson

def poisson_pmf(k, mu):
    """
    Calculates the probability mass function (PMF) of a Poisson distribution.

    Args:
        k: The number of events.
        mu: The average rate of events.

    Returns:
        The PMF value.
    """
    pmf = poisson.pmf(k, mu)
    return pmf
````

11. A company wants to test if a new website layout leads to a higher conversion rate (percentage of visitors
who make a purchase). They collect data from the old and new layouts to compare.


To generate the data use the following command:

```python

import numpy as np

# 50 purchases out of 1000 visitors

old_layout = np.array([1] * 50 + [0] * 950)

# 70 purchases out of 1000 visitors  

new_layout = np.array([1] * 70 + [0] * 930)

  ```

Apply z-test to find which layout is successful.

````python
import numpy as np
from scipy.stats import norm

# Data
old_layout = np.array([1] * 50 + [0] * 950)
new_layout = np.array([1] * 70 + [0] * 930)

# Calculate proportions
old_proportion = np.mean(old_layout)
new_proportion = np.mean(new_layout)

# Calculate the pooled proportion
pooled_proportion = (np.sum(old_layout) + np.sum(new_layout)) / (len(old_layout) + len(new_layout))

# Calculate the standard error
se = np.sqrt(pooled_proportion * (1 - pooled_proportion) * (1 / len(old_layout) + 1 / len(new_layout)))

# Calculate the z-score
z_score = (new_proportion - old_proportion) / se

# Calculate the p-value
p_value = 2 * (1 - norm.cdf(abs(z_score)))

print(f"Z-score: {z_score}")
print(f"P-value: {p_value}")

# Interpret the results
alpha = 0.05  # Significance level

if p_value < alpha:
    print("Reject the null hypothesis. There is a statistically significant difference in conversion rates.")
    if new_proportion > old_proportion:
        print("The new layout is more successful.")
    else:
        print("The old layout is more successful.")
else:
    print("Fail to reject the null hypothesis. There is no statistically significant difference in conversion rates.")
````

```python
after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])

```

Use z-test to find if the claims made by tutor are true or false.

````python
from scipy.stats import norm

# Data
before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])
after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])

# Calculate differences in scores
differences = after_program - before_program

# Calculate mean and standard deviation of the differences
mean_difference = np.mean(differences)
std_difference = np.std(differences, ddof=1)  # Use ddof=1 for sample standard deviation

# Calculate the t-statistic
t_statistic = mean_difference / (std_difference / np.sqrt(len(differences)))

# Degrees of freedom
df = len(differences) - 1

# Calculate the p-value (two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(t_statistic)))


print(f"Mean Difference: {mean_difference}")
print(f"Standard Deviation of Differences: {std_difference}")
print(f"t-statistic: {t_statistic}")
print(f"p-value: {p_value}")

# Interpret the results
alpha = 0.05  # Significance level

if p_value < alpha:
    print("Reject the null hypothesis. The tutoring program significantly improves exam scores.")
else:
    print("Fail to reject the null hypothesis. There is no significant evidence that the tutoring program improves exam scores.")
````
